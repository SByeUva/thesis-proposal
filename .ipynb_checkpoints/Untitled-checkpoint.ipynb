```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as data
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
```

```python
# S&P 500, Dow jones and IQ CPI inflation tracker
tickers = ['^GSPC', '^DJI']

# Time line
start_date = '1990-10-23'
end_date = '2022-10-23'
```

```python
# Read in data
sp_raw = data.DataReader(tickers[0], 'yahoo', start_date, end_date)
dj_raw = data.DataReader(tickers[1], 'yahoo', start_date, end_date)
```

```python
# Subset to monthly data
sp_month = sp_raw.iloc[::24, :]
dj_month = dj_raw.iloc[::24, :]
```

```python
# Generate monthly difference data
diff_sp = np.diff(np.log(sp_month['Close']))
diff_dj = np.diff(np.log(dj_month['Close']))

# Set difference data to df
diff_sp_m = np.zeros(len(diff_sp) + 1)
diff_sp_m[0:-1] = diff_sp
sp_month['diff'] = diff_sp_m

diff_dj_m = np.zeros(len(diff_dj) + 1)
diff_dj_m[0:-1] = diff_dj
dj_month['diff'] = diff_dj_m
```

```python
fig, axs = plt.subplots(2,2, figsize=(15,15))
plt.rcParams.update({'font.size': 10})

axs[0,0].plot(sp_month['diff'])
axs[0,1].plot(dj_month['diff'])
sns.histplot(sp_month['diff'], ax = axs[1,0])
sns.histplot(dj_month['diff'], ax = axs[1,1])

axs[0,0].set_ylabel('$\log(S_{t+1} - S_{t})$')
axs[0,0].set_title('Log difference of montly S&P500')
axs[0,1].set_title('Log difference of montly DJI')

plt.show()
```

```python
# Distribution fit tests
print(stats.shapiro(sp_month['diff']))
print(stats.shapiro(dj_month['diff']))
print(stats.kstest(sp_month['diff'], 'norm', 
                   args = (np.mean(sp_month['diff']), np.std(sp_month['diff']))))
print(stats.kstest(dj_month['diff'], 'norm',
                   args = (np.mean(dj_month['diff']), np.std(dj_month['diff']))))
```

```python
sp_splits = np.array_split(sp_month['diff'].to_numpy(), 6)
dj_splits = np.array_split(dj_month['diff'].to_numpy(), 6)
```

```python
for split in sp_splits:
    print(
        stats.kstest(split, 'norm', 
             args = (np.mean(split), np.std(split)))
    )
```

```python
for split in dj_splits:
    print(
        stats.kstest(split, 'norm', 
             args = (np.mean(split), np.std(split)))
    )
```
